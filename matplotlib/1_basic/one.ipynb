```python
import pandas as pd
import matplotlib.pyplot as plt

air_quality = pd.read_csv("../data/air_quality_no2.csv", index_col=0, parse_dates=True)
air_quality.head()
```

```python
air_quality['station_paris'].plot()
```

```python
# I want to visually compare the values measured in London versus Paris.

air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.3)
plt.show()
```

```python
# Name of the methods that are exists

[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

```python
air_quality.plot.area(x="station_london", y="station_paris", stacked=True,alpha=0.3)
```

```python
air_quality.plot.box()
```

```python
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
axs
```
